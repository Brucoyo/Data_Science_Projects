```
# Question 1
wordsRDD = sc.textFile("/FileStore/tables/q1.txt")
output = sc.parallelize(wordsRDD.collect())\
.flatMap(lambda line: line.split(","))\
.map(lambda word: [word,(word+"es")])
print(output.collect())
```

```
# Question 2
raw_data = sc.textFile("/FileStore/tables/q2.csv")
lines = sc.parallelize(raw_data.collect())
text=lines.map(lambda line: line.split(","))

# 2.3
print("Question 2.3")
A3=[lines.count(),lines.take(1)]
print(A3)

# 2.4
print("Question 2.4")
print(sc.parallelize(text.collect())\
.map(lambda year: year[0]).distinct().count())

# 2.5
print("Question 2.5")
A5= sc.parallelize(text.collect())\
.map(lambda group: (group[1],1))\
.reduceByKey(lambda a, b: a + b)
print("without sorting \n",A5.collect())
```

```
# 2.6
print("Question 2.6")
A6= sc.parallelize(text.collect())\
.map(lambda group: (group[1],1))\
.reduceByKey(lambda a, b: a + b)\
.sortByKey(ascending=False)
print("with descending sorting \n",A6.collect())
```

```
# 2.7
print("Question 2.7")
A7 = sc.parallelize(text.collect())\
.filter(lambda x: x[4]=="President Barack Obama")
print("how many times  President Barack Obama appeared on ‘The Daily Show’?",A7.count())
```

```
# Question 3
rdd_data = sc.textFile("/FileStore/tables/q3.txt")
#rdd_data.collect()
results_line=rdd_data.map(lambda result: result.split("\t"))

# Q 3.2
# remove header
print("Q 3.2")
data_header = results_line.first()
#header.collect()
data = results_line.filter(lambda line: line != data_header)
print("number of entries stored into the data RDD:",data.count())

# 3.3
print("Q 3.3")
Q3_3 = sc.parallelize(data.collect())\
.filter(lambda x: int(x[2])>20)
print("number of cars that have MPG more than 20:",Q3_3.count())

# 3.4
print("Q 3.3")
A=sc.parallelize(data.collect())\
.filter(lambda x: int(x[3])>4)\
.filter(lambda car: "chevrolet" in car[0])
print("Count the number of  chevrolet car that has Cylinders more than 4:", A.count())
```
