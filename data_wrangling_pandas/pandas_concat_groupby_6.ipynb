```python
s1 = pd.Series([0, 1], index=['a', 'b'])
s2 = pd.Series([2, 3, 4], index=['c','d', 'e'])
s3 = pd.Series([5, 6], index=['f', 'g'])
s1
```

```python
pd.concat([s1,s2,s3])
```

```python
pd.concat([s1, s2, s3],axis=1, sort=False)
```

```python
# if axis =1 then add to a new column
# sort will sort the index by order
s1 = pd.Series([0, 1], index=['a', 'b'])
s4 = pd.Series([0, 1, 100, 6],index=['a','b','h', 'g'])
pd.concat([s1, s4], axis=1, sort=True)
```

```python
# if axis =0, then append to the bottom row
pd.concat([s1, s4], axis=0, sort=True)
```

```python
# join_axes choose index in the new table
pd.concat([s1, s4], axis=1,join_axes=[['a', 'c', 'b', 'e','f','zz']])
```

```python
# form a hierarchical index along the concatenation axis
result = pd.concat([s1, s1, s3], keys=['one','two', 'three'])
result
```

```python
result.unstack(level=0)
```

```python
bb=pd.concat([s1, s1, s3], keys=['one', 'two','three'], axis = 1)
bb
```

## concat vs append

```python
# ***
pd.concat([df1,df2],axis=1,keys=['level1','level2'],sort=True,names=['upper','lower'])
```

```python
# join_axes=[df1.columns] will join df1.columns ignoring df2's
# even duplicate indices
# use verify_integrity attribute to catch errors

try:
    a=pd.concat([df1, df2],
    join_axes=[df1.columns],
    verify_integrity=True,sort = True)
except ValueError as e:
    print("ValueError:", e)
except:
    print("error")
a
```

```python
# add rows (df2,s1,s2) over df1
df1.append([df2,s1,s2])
```

```python
pd.concat([df1,df2,s1],join_axes=[df1.columns],ignore_index=False).tail()
```

```python
pd.concat([df1,df2,s1],join_axes=[df1.columns],ignore_index=True).tail()
```

```python
# this will combine the data of df2 to df1 first then
# combine the table
df1.combine_first(df2)
```

```python
# specify the axes to be used on other axes with join_axes
pd.concat([s1,s4],axis=1,join_axes=[['a','c','b','e']])
```

### merge (tought this first in class)

```python
import pandas as pd
```

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
```

```python
right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
df6 = pd.DataFrame({'name': ['Peter','Paul',"Mary"] ,
                     'food': ['fish','beans','bread']})
df7 = pd.DataFrame({'name': ['Mary','Joseph'] ,
                     'drink': ['wine','beer']})
```

```python
df6
```

```python
df7
```

```python
# intersection
pd.merge(df6,df7)
```

```python
# left join
pd.merge(df6,df7,how='left')
```

```python
# right join
pd.merge(df6,df7,how='right')
```

```python
# inner join (intersection) # set by default
pd.merge(df6,df7,how='inner')
```

```python
# outer join (union)
pd.merge(df6,df7,how='outer')
```

```python
# on if there are something in common
pd.merge(df6,df7, on = "name")
```

```python
left_df=pd.DataFrame({'key1': ['Ohio','Ohio',"Ohio","Nevada","Nevada"] ,
                     'key2': ['2000','2001','2002','2001','2002'] ,
                      'data': [0,1,2,3,4]})

index = pd.MultiIndex.from_tuples([('Nevada', '2001'), ('Nevada', '2000'),
                                   ('Ohio', '2000'), ('Ohio', '2000'),
                                  ("Ohio",'2001'), ('Ohio','2002')])

## how to make multiline index ? fixed!
right_df=pd.DataFrame({'event1':[0,2,4,6,8,10],
                       'event2':[1,3,5,7,9,11]},
                     index=index)
```

```python
left_df
```

```python
right_df
```

```python
pd.merge(left_df, right_df, left_on=['key1','key2'],right_index=True,how='outer')
```

```python
pd.merge(left_df, right_df, left_on=['key1','key2'],right_index=True,how='inner')
```
