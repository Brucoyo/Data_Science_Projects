```python
import pandas as pd
```

```python
# read dataset and create the pivot table
employee = pd.read_csv('employee.csv')
employee = employee.pivot_table(values='Value',
                            index=['DATE'],
                            columns=['UNIT','AGE', 'SEX', 'INDIC_EM', 'GEO'])
employee.index = pd.to_datetime(employee.index) # ensure that dates are datetime format
employee.head()
```

```python
# You need to filter by country, so, swap GEO to the top level 
employee.columns = employee.columns.swaplevel(0,-1)

# Now sort the MultiIndex
employee = employee.sort_index(axis=1)

# From the dataset, you know that not all columns are countries. which ones?
# We need to eliminate them in GEO: the items that are not countries
# Create a list to find the level values in GEO that begin with â€˜Euroâ€™
geo_list = employee.columns.get_level_values('GEO').unique().tolist()
countries = [x for x in geo_list if not x.startswith('Euro')]
employee = employee[countries]

# Now you can get the level values of the columns in employee
employee.columns.get_level_values('GEO').unique()
```

```python
# Your one line code goes here. Hint: use pandas function xs()
e1=employee.xs(("Percentage of total population","Active population"),axis=1,level=(4,3))
e1.head()
```

```python
# now drop the 'Total' column in the 'SEX' level 
#to have a more condensed form
e2=e1.drop("Total",axis=1,level=2)
e2.head()
```

One task of interest might be to compute a DataFrame consisting of the yearly correlations of daily returns (computed from percent changes) with SPX . 

As one way to do this, we first create a function that computes the pairwise correlation of each column with the 'SPX'  column. Use 'corrwith' function. (10 Points)

```python
# SPX
close_px.corr(method="pearson")
```

```python
close_px.drop(labels="SPX",axis=1).corrwith(close_px["SPX"])
```

Next, we compute percent change on close_px  using 'pct_change' function: (10 Points)

```python
# complete the code
rets = close_px.pct_change()
rets.head()
```

Lastly, we group these percent changes by year, which can be extracted from each row label with a one-line function that returns the year  attribute of each datetime  label: (25 Points)

```python
rets.groupby(rets.index.to_period("Y")).agg("sum")
```
