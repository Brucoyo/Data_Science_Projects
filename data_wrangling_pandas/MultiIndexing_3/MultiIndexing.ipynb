# Q 6

```python
import numpy as np
Q6=pd.pivot_table(df, values=["Value",],index=["GEO","AGE","SEX","DATE"],columns=["UNIT","INDIC_EM",],aggfunc=np.mean,fill_value=0)
Q6
```

7. Write a program that returns all values in the MultiIndex (name of inner levels). For example, for Age, it should return 'From 15 to 24 years', etc. 

```python
print("name of inner levels in the multiindex \n")
print("GEO",Q6.index.get_level_values(0).unique(),"\n")
print("AGE",Q6.index.get_level_values(1).unique(),"\n")
print("SEX", Q6.index.get_level_values(2).unique(),"\n")
print("DATE", Q6.index.get_level_values(3).unique(),"\n")

#print("level 0",Q6.index.levels[0].tolist(),"\n")
#print("level 1",Q6.index.levels[1].tolist(),"\n")
#print("level 2",Q6.index.levels[2].tolist(),"\n")
#print("level 3",Q6.index.levels[3].tolist(),"\n")
#print("level 4",Q6.index.levels[4].tolist(),"\n")
```
