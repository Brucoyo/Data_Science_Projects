## Q1

```python
# code
import csv
with open('Bison-observation-2002.csv','r',encoding='utf-8') as f:
    reader = csv.reader(f,delimiter=',')
    Bison_count = int(0)

    next(reader)
    next(reader)

    for row in reader:
        try:
            Bison_count += int(row[7])
        except:
            pass        
print("The sum of the Bisons observed in the dataset",Bison_count)
```

## Q 2

```python
import csv
import numpy as np
with open('Bison-observation-2002.csv','r',encoding='utf-8') as f:
    reader = csv.reader(f,delimiter=',')
    
    sub_total=[]
    sub1=[]
    
    next(reader)
    next(reader)
    
    for row in reader:
        sub_total.append(row[1])
        if row[1] not in sub1:
            sub1.append(row[1])

print("Number of different types of Bisons are observed in the 'Sub Population' column:",len(sub1))
```
